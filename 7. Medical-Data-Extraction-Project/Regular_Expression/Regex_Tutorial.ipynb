```python
import re

text = '''
Name: Marta Sharapova Date: 5/11/2022

Address: 9 tennis court, new Russia, DC

Prednisone 20 mg
Lialda 2.4 gram

Directions:

Prednisone, Taper 5 mig every 3 days,
Finish in 2.5 weeks a
Lialda - take 2 pill everyday for 1 month

Refill: 2 times
'''

# 1. Name
name = re.findall(r"Name:(.*)Date", text)[0].strip()

# 2. Date
date = re.findall(r"Date:(.*)\n", text)[0].strip()

# 3. Address
address = re.findall(r"Address:(.*)\n", text)[0].strip()

# 4. Medicines (lines between Address and Directions)
med_pattern = r"Address[^\n]*\n(.*)Directions"
medicines = re.findall(med_pattern, text, flags=re.DOTALL)[0].strip()

# 5. Directions (multi-line block)
dir_pattern = r"Directions:(.*)Refill"
directions = re.findall(dir_pattern, text, flags=re.DOTALL)[0].strip()

# 6. Refill
refill = re.findall(r"Refill:(.*)times", text)[0].strip()


# -------- PRINT OUTPUT --------
print("Name:", name)
print("Date:", date)
print("Address:", address)
print("\nMedicines:\n", medicines)
print("\nDirections:\n", directions)
print("\nRefill:", refill)
```
